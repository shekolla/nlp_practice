```python
with open('friends-tv-show-script/Friends_Transcript.txt') as f:
        text = f.read()
```

```python
len(text)
len(text.split('\n'))
```

```python
tokens = text.split('\n')
tokens_n = list(map(lambda x: simple_preprocess(x), tokens))
```

```python
print(len(tokens_n[0]))
```

```python
wv = Word2Vec(window=10, min_count=2)
```

```python
wv.build_vocab(tokens_n)
```

```python
wv.train(tokens_n, total_examples=wv.corpus_count, epochs=wv.epochs)
```

```python
wv.wv.most_similar('ross')
```

```python
!ls game-of-thrones-book-dataset/
```

```python
got_text = []
for filename in os.listdir('game-of-thrones-book-dataset/'):
    file_path = os.path.join('game-of-thrones-book-dataset/', filename)
    with open(file_path, 'r', encoding='latin1') as f:
        text_book = f.read()
    for each_sen in sent_tokenize(text_book):
        got_text.append(simple_preprocess(each_sen))

    
```

```python
got_wv = gensim.models.Word2Vec(window=20)
```

```python
got_wv.build_vocab(got_text)
```

```python
got_wv.train(got_text, total_examples=got_wv.corpus_count, epochs=got_wv.epochs)
```

```python
got_wv.wv.index_to_key
```

```python
got_wv.wv.most_similar('smallfolk')
```
